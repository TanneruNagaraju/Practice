```python
N = 100
D = 2
```

```python
X = np.random.rand(N,D)
```

```python
#first 50 rows are subtracted with 2 value 
X[:50,:] = X[:50,:] - 2*np.ones((50,D))
```

```python
X[:50,:]
```

```python
print(X)
print(X.shape)
```

```python
#last 50 rows are added with 2 value 

X[50:,:] = X[50:,:] + 2*np.ones((50,D))  # last 50 rows are added with 2 value 
```

```python
print(X)
```

```python
# first 50 rows are 0's & last 50 rows are 1's 
T = np.array([0]*50+[1]*50) # T= Targets
print(T)  
              
```

```python
ones = np.ones((N,1))
print(ones)
```

```python
Xb = np.concatenate((X,ones),axis = 1) #  x is input & ones is bias
```

```python
print(Xb)
```

```python
# weights
w = np.random.randn(D+1) #randomly initialise weights
```

```python
print(w)
```

```python
#predictions
Z  = np.dot(Xb,w)
print(Z)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-Z))
```

```python
Y = sigmoid(Z)
```

```python
print(Y)
```

```python
def cross_entropy_error(T,Y):
    E = 0
    for i in range(N):
        if T[i] == 1:
            E-=np.log(Y[i])
        else:
            E-=np.log(1 - Y[i])
    return E        
            
```

```python
cross_entropy_error(T,Y)
```

```python
w = np.array([0,4,4]) #bias,two columns values 4,4
```

```python
Z = np.dot(Xb,w)
print(Z)
```
