```python
import numpy as np
```

```python
N = 50
D = 2 
```

```python
X = np.random.rand(N,D)
ones = np.ones((N,1))
```

```python
X
```

```python
ones
```

```python
xones = np.concatenate((ones,X),axis = 1)
```

```python
xones
```

```python
w = np.random.randn(D+1)
```

```python
w
```

```python
Y = np.dot(xones,w)
```

```python
print(Y)
```

```python
sigmoid = 1/1+np.exp(-Y)
```

```python
print(sigmoid)
```
