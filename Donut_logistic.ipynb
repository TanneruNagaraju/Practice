```python
Xb = np.concatenate((ones,r,X),axis = 1)
```

```python
print(Xb)
```

```python
Xb.shape
```

```python
w = np.random.rand(D+2) # four weights for 4 columns
```

```python
w
```

```python
Z = np.dot(Xb,w)
```

```python
print(Z)
```

```python
def sigmoid(Z):
    return 1/(1+np.exp(-Z))
```

```python
Y = sigmoid(Z)
```

```python
print(Y)
```

```python
def cross_entropy(T,Y):
    return -(T*np.log(Y)+(1-T)*np.log(1-Y)).sum()
```

```python
cross_entropy(T,Y)
```

```python
learning_rate = 0.001
```

```python
error = []
```

```python
for i in range(5000):
    e = cross_entropy(T,Y)
    error.append(e)
    if i % 500 == 0:
        print(e) 
        # l2 regularisation
    w = w + learning_rate*(Xb.T.dot(T-Y)-0.01*w) #l2 = 0.01
    
    Y  = sigmoid(Xb.dot(w))
```

```python
plt.plot(error)
plt.title("Cross_entropy per iteration")
plt.show()
```

```python
print("final",w)
```

```python
print("final classification rate:",1 - np.abs(T-np.round(Y)).sum()/N)
```
