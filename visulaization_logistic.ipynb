```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
D = 2
```

```python
X = np.random.rand(N,D)
```

```python
print(X)
```

```python
X[:50,:] = X[:50,:] - 2*np.ones((50,D))
```

```python
print(X)
```

```python
X[50:,:] = X[50:,:] + 2*np.ones((50,D))
```

```python
print(X)
```

```python
# Targets
T = np.array([0]*50 + [1]*50)
print(T)
```

```python
ones = np.ones((N,1))
```

```python
ones
```

```python
Xb = np.concatenate((X,ones),axis = 1)
```

```python
print(Xb)
```

```python
w = np.array([0,4,4])
print(w)
```

```python
Z =  np.dot(Xb,w)
print(Z)
```

```python
def sigmoid(Z):
    return 1/(1+np.exp(-Z))
```

```python
Y = sigmoid(Z)
```

```python
print(Y)
```
