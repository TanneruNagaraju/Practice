```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50
D = 50
```

```python
X = (np.random.random((N,D))-0.5)*10
```

```python
np.random.random((N,D))
```

```python
np.random.random((N,D))-0.5
```

```python
w1 =  np.array([1.0,0.5,-0.5]+[0]*(D-3)) # first 3 rows are 1.0,0.5,-0.5 remaining 47 values are zero
```

```python
print(w1)
```

```python
len(w1)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
sigmoid(np.dot(X,w1))
```

```python
np.random.randn(N)*0.5
```

```python
Y =  np.round(sigmoid(np.dot(X,w1)+np.random.randn(N)*0.5))
```

```python
print(Y)
```

```python
costs = []
```

```python
learning_rate = 0.001
```

```python
w = np.random.randn(D)/np.sqrt(D)
```

```python
print(w)
```

```python
np.sqrt(D)
```

```python
l1 = 3.0
```

```python
np.abs(w) #converting negative values to positive values
```

```python
sigmoid(np.dot(X,w))
```

```python
for t in range(1000):
    Ypred = sigmoid(np.dot(X,w))
    delta = Ypred - Y
    w = w -  learning_rate*(np.dot(X,delta)+l1*np.sign(w))
    cost = -(Y*np.log(Ypred)+(1-Y)*np.log(Ypred)).mean()+l1*np.abs(w).mean()
    costs.append(cost)
    
    
```

```python
costs
```

```python
plt.plot(costs)
plt.show()
```

```python
print("final",w)
```
