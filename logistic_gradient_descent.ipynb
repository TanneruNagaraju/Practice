```python
import numpy as np
```

```python
N = 100
D = 2
```

```python
X = np.random.randn(N,D)
print(X)
```

```python
X[:50,:] = X[:50,:] - 2*np.ones((50,D))
```

```python
print(X)
```

```python
X[50:,:] = X[50:,:] + 2*np.ones((50,D))
```

```python
print(X)
```

```python
T = np.array([0]*50+[1]*50)
print(T)
```

```python
ones = np.ones((N,1))
print(ones)
```

```python
Xb = np.concatenate((X,ones),axis = 1)
print(Xb)
```

```python
w = np.random.randn(D+1)
print(w)
```

```python
z = np.dot(Xb,w)
print(z)
```

```python
sigmoid = 1/(1+np.exp(-z))
```

```python
print(sigmoid)
```

```python
Y = sigmoid
print(Y)
```

```python
def cross_entropy_error(T,Y):
    E = 0
    for i in range(N):
        if T[i]==1:
            E-=np.log(Y[i])
        else:
            E-=np.log(1-Y[i])
    return E        
```

```python
cross_entropy_error(T,Y)
```
