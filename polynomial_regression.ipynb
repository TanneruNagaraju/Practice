```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = []
Y = []
```

```python
for i in open("C:/Users/TANNERU/Downloads/Dataset for coding/data_poly.csv"):
    x,y = i.split(',')
    #print(x,y)
    x = float(x)
    y = float(y)
    X.append([1,x,x*x]) # x^2+x+c
    Y.append(y)
    
```

```python
print(X)
print(Y)
```

```python
X = np.array(X)
Y = np.array(Y)
print(X)
print(Y)
```

```python
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[:,1],Y)
plt.show()
```

```python
weights = np.linalg.solve(np.dot(X.T,X),np.dot(X.T,Y))
Ypred = np.dot(X,weights)
```

```python
print(weights)
```

```python
print(Ypred)
```

```python
print(Ypred.size)
```

```python
print(Ypred.dtype)
```

```python
compare = pd.DataFrame({"Actual":Y,"predict":Ypred})
print(compare)
```

```python
plt.scatter(X[:,1],Y)
plt.plot(sorted(X[:,1]),sorted(Ypred))
plt.show()
```

```python
ss = Y - Ypred
sr = Y - Y.mean()
```

```python
print(ss)
```

```python
print(sr)
```

```python
r2_score = 1 - np.dot(ss,ss)/np.dot(sr,sr)
print("R2_score",r2_score)
```

```python
from sklearn.metrics import r2_score
print("R2_score:",r2_score(Y,Ypred))
```
