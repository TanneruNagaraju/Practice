```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
X = []
Y = []
```

```python
for i in open("C:/Users/TANNERU/Downloads/Dataset for coding/data_2d.csv"):
    x1,x2,y = i.split(',')
    #print(x1,x2,y)
    X.append([float(x1),float(x2),1]) # 1 is bias term 
    Y.append(float(y))
```

```python
print(X)
print(Y)
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
print(X)
print(Y)
```

```python
print(X.shape)
print(Y.shape)
```

```python
fig = plt.figure()
axes = fig.add_subplot(111,projection = '3d')
axes.scatter(X[:,0],X[:,1],Y)
plt.show()
```

```python
weights = np.linalg.solve(np.dot(X.T,X),np.dot(X.T,Y))
Ypred = np.dot(X,weights)
```

```python
print(weights)
```

```python
print(Ypred)
```

```python
import pandas as pd
```

```python
compare = pd.DataFrame({'Actual':Y,'Predict':Ypred})
print(compare)
```

```python
ss = Y -Ypred
sr = Y- Y.mean()
print(ss)
print(sr)
```

```python
#R2_score
r2_score = 1 - np.dot(ss,ss)/np.dot(sr,sr)

print("R2_score:",r2_score)
```

```python
from sklearn.metrics import r2_score
print("R2_score:",r2_score(Y,Ypred))
```
